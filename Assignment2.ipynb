# Task 1

Iris Dataset is prominently known for simple machine learning problems. It consits of three species of Iris which are: Iris setosa, Iris virginica and Iris versicolor; and each having 50 samples (observations). Hence, a total of 150 sample or observation.

Features measure from the samples are the length of sepals, the width of sepals, the length of petals and width of petals. The metric used is centimeters.

The main dataset is divided into two which are Iris_data and Iris_labels. The features are sepal_length, sepal_width, petal_length, and petal_width.

Iris_label contain the classes of the Iris which has been encoded as o,1 and 2

```python
#importing the libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
#loading the dataset
data = np.load('Iris_data.npy')
labels = np.load('Iris_labels.npy')
```

```python
xFeatures = 0
yFeatures = 1

x = data[:,xFeatures]
y = data[:, yFeatures]

colours = labels

plt.scatter(x, y, c=colours, cmap='viridis')

plt.xlabel('Iris Observations')
plt.ylabel('Iris Classes')
plt.title('Visualization of Fisher Iris Dataset')

#labels=["Iris Setosa", "Iris Versicolour", "Iris Virginica"]
#plt.legend()

plt.show()
```

# Task 2

```python
from sklearn.cluster import KMeans
```

```python
X = data
kmeans = KMeans(n_clusters=3, random_state=1).fit(X)
```

```python
print(kmeans.labels_)
```

```python
kmeans.predict([[5.0,3.6,1.4,0.2], [5.0,2.0,3.5,1.0], [6.3,3.4,5.6,2.4]])
```

```python
print(kmeans.cluster_centers_)
```

```python
# Plot the data
plt.scatter(X[labels==0, 0], X[labels==0, 1])
plt.scatter(X[labels==1, 0], X[labels==1, 1])
plt.scatter(X[labels==2, 0], X[labels==2, 1])


# Plot the clusters
plt.scatter(kmeans.cluster_centers_[:, xFeatures], 
            kmeans.cluster_centers_[:, yFeatures],
            c='r', marker='X', s=100)

plt.xlabel('Iris Observations')
plt.ylabel('Iris Classes')
plt.title('Visualization of Fisher Iris Dataset (K-Means Model)')

plt.show()
```

# Task 3

```python
from sklearn.mixture import GaussianMixture
```

```python
X = data
gm = GaussianMixture(n_components=3, random_state=0).fit(X)
gm.means_
```

```python
gm.predict([[5.0,3.6,1.4,0.2], [5.0,2.0,3.5,1.0], [6.3,3.4,5.6,2.4]])
```

```python
#predictions from gm
labels = gm.predict(X)

colours = labels
plt.scatter(data[:,xFeatures],data[:,yFeatures],c=colours, cmap='viridis')

plt.xlabel('Iris Observations')
plt.ylabel('Iris Classes')
plt.title('Visualization of Fisher Iris Dataset (Gaussian Mixture Model)')

plt.show()
```
